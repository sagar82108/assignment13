```python
"""ans1) MongoDB is a popular document-oriented NoSQL (non-relational) database system. Non-relational databases, also known as NoSQL databases,
are designed to handle unstructured or semi-structured data in a flexible and scalable manner.In MongoDB, data is stored in flexible JSON-like documents called BSON (Binary JSON),
which can contain various data types and structures.

There are several scenarios where MongoDB is preferred over SQL databases:

Flexible Schema: If your application deals with rapidly changing or evolving data structures,
                MongoDB's flexible schema allows you to easily adapt to these changes without requiring strict schema migrations.

Scalability and Performance: MongoDB is designed to scale horizontally across multiple servers, 
                             making it suitable for handling large volumes of data and high traffic loads. 
                             It offers automatic sharding and built-in replication features for distributing data and ensuring high availability.

Big Data and Real-Time Analytics: MongoDB's flexible data model and powerful querying capabilities make it well-suited for storing and analyzing large volumes of unstructured or semi-structured data, such as log files, social media feeds, or sensor data.

Rapid Prototyping and Agile Development: MongoDB's flexible schema and ease of use allow for quicker iterations and prototyping during the development process. 
                                        It can be particularly useful in agile development environments where requirements may change frequently.
```

```python
"""ans2) MongoDB, as a document-oriented NoSQL database, offers several key features that contribute to its popularity and versatility. 
         Here are some of the notable features of MongoDB:

Document Model: MongoDB stores data in flexible, self-describing documents called BSON (Binary JSON).
                Documents are similar to JSON objects and can have varying structures and data types within the same collection. 
                This schema-less approach allows for easy handling of evolving data structures.

Scalability and High Availability: MongoDB is designed to scale horizontally by distributing data across multiple servers or clusters. 
                                   It supports automatic sharding, which divides data and distributes it across shards for improved performance and scalability.
                                   MongoDB also provides built-in replication for high availability, allowing for failover and data redundancy.

Querying and Indexing: MongoDB offers a powerful and expressive query language for retrieving data.
                       It supports a wide range of querying operations, including filtering, sorting, and aggregation. 
                      Indexes can be created on fields to optimize query performance. MongoDB also provides a rich set of operators and functions for complex data manipulation and analysis.

Flexible Data Model: The document-based data model of MongoDB allows for dynamic and nested data structures.
                    Documents can contain arrays, sub-documents, and other complex data types. 
                    This flexibility makes MongoDB suitable for handling complex, hierarchical, or evolving data.

Replication and Data Consistency: MongoDB supports replica sets, which are self-healing clusters that provide data redundancy and automatic failover. 
                                  Replica sets ensure high availability and durability of data. MongoDB also offers configurable levels of data consistency,
                                 allowing developers to choose the appropriate trade-off between consistency and performance.

Aggregation Framework: MongoDB includes a powerful aggregation framework that allows for advanced data processing and analytics.
                       It supports various operations like grouping, sorting, filtering, and transformation of data within the database.
                      The aggregation framework can efficiently process large volumes of data and perform complex computations.

Geospatial and Full-Text Search: MongoDB has built-in support for geospatial data and provides geospatial indexing and querying capabilities. 
                                 It enables applications to store and query location-based data effectively. 
                                 MongoDB also offers full-text search functionality for text-based searching and indexing of textual content.

Easy Integration and Development: MongoDB provides official drivers and libraries for various programming languages,
                                  making it easy to integrate with different application stacks. It offers a flexible query API and provides rich data manipulation features,
                                  facilitating rapid application development and prototyping.






```

```python
"""ans3) 

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017')

# Create a database
database = client['mydatabase']

# Create a collection
collection = database['mycollection']

# Test data
data = {'name': 'John', 'age': 30, 'city': 'New York'}

# Insert a document into the collection
collection.insert_one(data)

# Find all documents in the collection
cursor = collection.find()

# Print each document
for document in cursor:
    print(document)
```

```python
"""ans4) 

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017')

# Access the database and collection
database = client['mydatabase']
collection = database['mycollection']

# Insert one record
data_one = {'name': 'Alice', 'age': 25, 'city': 'London'}
collection.insert_one(data_one)

# Insert many records
data_many = [
    {'name': 'Bob', 'age': 32, 'city': 'Paris'},
    {'name': 'Charlie', 'age': 28, 'city': 'Berlin'},
    {'name': 'David', 'age': 35, 'city': 'Tokyo'}
]
collection.insert_many(data_many)

# Print inserted record using find_one()
print("One inserted record:")
print(collection.find_one())

# Print all inserted records using find()
print("All inserted records:")
cursor = collection.find()
for document in cursor:
    print(document)
```
